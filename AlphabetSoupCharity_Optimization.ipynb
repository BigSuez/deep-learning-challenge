## Initial Preprocessing

```python
# Import dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop unnecessary columns
application_df = application_df.drop(columns=['EIN', 'NAME'], axis=1)
```

## Optimization Attempt #1
- Reducing Application Type and Classification Bins
- Reduced to 1 Hidden Layer

```python
#Create New DataFrame for Optimization Attempt #1
optim_1_df = application_df.copy()
```

```python
#Binning Applicaiton Types
application_types_to_replace = ['T9', 'T13', 'T12', 'T2', 'T25', 'T14', 'T29', 'T15', 'T17', 'T10']
# Replace in dataframe
for app in application_types_to_replace:
    optim_1_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")
```

```python
#Binning Classifications
classifications_to_replace = list(application_df['CLASSIFICATION'].unique())
classifications_to_keep = ['C1000', 'C2000', 'C1200', 'C3000', 'C2100', 'C7000']
for cls in classifications_to_keep:
    classifications_to_replace.remove(cls)

# Replace in dataframe
for cls in classifications_to_replace:
    optim_1_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
```

```python
# Get Dummy Columns
categorical_cols = list(optim_1_df.columns)
categorical_cols.remove('ASK_AMT')
processed_df = pd.get_dummies(optim_1_df[categorical_cols])
processed_df['ASK_AMT'] = optim_1_df['ASK_AMT']
processed_df.head()
```

```python
# Splitting Data
target = processed_df['IS_SUCCESSFUL']
features = processed_df.drop(['IS_SUCCESSFUL'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(features, target)
```

```python
#Scaling Data
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=8, activation='relu', input_dim=116))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Optimization Attempt #3
- Using Keras Tuner to find Best Model
- Input Data in same format as Unoptimized Model

```python
#Create New DataFrame for Optimization Attempt #1
optim_3_df = application_df.copy()
```

```python
#Binning Applicaiton Types
application_types_to_replace = ['T9', 'T13', 'T12', 'T2', 'T25', 'T14', 'T29', 'T15', 'T17']
# Replace in dataframe
for app in application_types_to_replace:
    optim_3_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")
```

```python
#Binning Classifications
classifications_to_replace = list(application_df['CLASSIFICATION'].unique())
classifications_to_keep = ['C1000', 'C2000', 'C1200', 'C3000', 'C2100']
for cls in classifications_to_keep:
    classifications_to_replace.remove(cls)

# Replace in dataframe
for cls in classifications_to_replace:
    optim_3_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
```

```python
# Get Dummy Columns
categorical_cols = list(optim_3_df.columns)
categorical_cols.remove('ASK_AMT')
processed_df = pd.get_dummies(optim_3_df[categorical_cols])
processed_df['ASK_AMT'] = optim_3_df['ASK_AMT']
processed_df.head()
```

```python
# Splitting Data
target = processed_df['IS_SUCCESSFUL']
features = processed_df.drop(['IS_SUCCESSFUL'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(features, target)
```

```python
#Scaling Data
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    model = tf.keras.models.Sequential()
    # Choice of Activation Functions
    activation = hp.Choice('activation', ['relu', 'tanh', 'sigmoid'])
    # Choice of first layer Neurons
    model.add(tf.keras.layers.Dense(units=hp.Int('first_unit',
                                                 min_value=1,
                                                 max_value=10,
                                                 step=2), activation=activation, input_dim=116))
    # Choice of Layers & Layer Neurons
    for i in range(hp.Int('num_layers', 1, 4)):
        model.add(tf.keras.layers.Dense(units=hp.Int('unit_'+str(i),
                                                     min_value=1,
                                                     max_value=10,
                                                     step=2), activation=activation))
    model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
    # Compile the model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python

# Import the kerastuner library
import keras_tuner as kt
# Create a `Hyperband()` tuner instance
tuner = kt.Hyperband(
    create_model,
    objective='val_accuracy',
    max_epochs=50,
    hyperband_iterations=2
)
```

```python
#Find top Models
tuner.search(X_train_scaled, y_train, epochs=20, validation_data=(X_test_scaled, y_test))
top_models = tuner.get_best_models(3)
```

```python
# Select the top model found
model = top_models[0]
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
